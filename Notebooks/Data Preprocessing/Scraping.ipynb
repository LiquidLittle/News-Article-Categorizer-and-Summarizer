```python
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
def huffpost_content(url):
    global count
    # Make an HTTP request to the webpage
    try: 
        response = requests.get(url)
    except:
        content = None
    else:
        # Parse the HTML contents of the page using beautifulsoup4
        soup = BeautifulSoup(response.content, "html.parser")

        content = ""
        for node in soup.find_all(class_="primary-cli cli cli-text"):
            content += node.get_text()
        
        print(count)
        count += 1

    return content
```

```python
for i in range(86,95):
    print("Starting: " + str(i) + " data")
    df_split[i]["content"] = df_split[i]["link"].apply(huffpost_content)
    df_split[i].to_csv("huffdata" + str(i) + ".csv")
    print("Finished: " + str(i) + " data")
```
