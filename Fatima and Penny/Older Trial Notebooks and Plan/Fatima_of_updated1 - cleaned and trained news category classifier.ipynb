```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def evaluate(model, dataloader, device):
    model.eval()  # Set the model to evaluation mode
    predictions, true_labels = [], []

    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            # Move logits and labels to CPU
            logits = logits.detach().cpu().numpy()
            label_ids = labels.to('cpu').numpy()

            # Store predictions and true labels
            predictions.extend(np.argmax(logits, axis=1).flatten())
            true_labels.extend(label_ids.flatten())

    # Calculate accuracy and other metrics
    accuracy = accuracy_score(true_labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(true_labels, predictions, average='weighted')

    return accuracy, precision, recall, f1
```

```python
import pandas as pd
import numpy as np
import torch
from transformers import BartTokenizer, BartForConditionalGeneration, AdamW
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split

# Load your dataset
path = '/content/drive/MyDrive/project/merge_df.csv'
data = pd.read_csv(path)

# Tokenization
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)

# Define a custom Dataset class
class SummaryDataset(Dataset):
    def __init__(self, tokenizer, articles, summaries, max_length=512, summary_length=128):
        self.input_ids = []
        self.attention_masks = []
        self.labels = []

        for article, summary in zip(articles, summaries):
            encoded_article = tokenizer.encode_plus(article, max_length=max_length,
                                                    padding='max_length', truncation=True,
                                                    return_tensors='pt')
            encoded_summary = tokenizer.encode_plus(summary, max_length=summary_length,
                                                    padding='max_length', truncation=True,
                                                    return_tensors='pt')

            self.input_ids.append(encoded_article['input_ids'].flatten())
            self.attention_masks.append(encoded_article['attention_mask'].flatten())
            self.labels.append(encoded_summary['input_ids'].flatten())

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_masks[idx],
            'labels': self.labels[idx]
        }

# Prepare the dataset
train_articles, val_articles, train_summaries, val_summaries = train_test_split(
    data['Articles'].tolist(), data['Summaries'].tolist(), test_size=0.2)

train_dataset = SummaryDataset(tokenizer, train_articles, train_summaries)
val_dataset = SummaryDataset(tokenizer, val_articles, val_summaries)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)

# Load the model
model = BartForConditionalGeneration.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Optimizer
optimizer = AdamW(model.parameters(), lr=5e-5)

# Training loop
for epoch in range(3):
    model.train()
    for batch in train_loader:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch + 1}: Loss {loss.item()}")

# Save the fine-tuned model
model.save_pretrained('/content/drive/MyDrive/project/fine_tuned_bart')

# Note: You can also add validation step to evaluate the model's performance on the val_loader dataset
```

```python
from rouge_score import rouge_scorer

# Function to evaluate the model using the ROUGE metric
def evaluate_model(model, dataloader, tokenizer, device):
    model.eval()
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            outputs = model.generate(input_ids=input_ids, attention_mask=attention_mask)

            for output, label in zip(outputs, batch['labels']):
                generated = tokenizer.decode(output, skip_special_tokens=True)
                reference = tokenizer.decode(label, skip_special_tokens=True)
                score = scorer.score(reference, generated)
                rouge_scores['rouge1'].append(score['rouge1'].fmeasure)
                rouge_scores['rouge2'].append(score['rouge2'].fmeasure)
                rouge_scores['rougeL'].append(score['rougeL'].fmeasure)

    # Calculate average scores
    avg_scores = {key: np.mean(val) for key, val in rouge_scores.items()}
    return avg_scores

# Evaluate the model
val_loader = DataLoader(val_dataset, batch_size=8)  # Assuming val_dataset is already defined
rouge_scores = evaluate_model(model, val_loader, tokenizer, device)
print("Average ROUGE scores:", rouge_scores)
```

```python
def evaluate_with_rouge(generated_summaries, reference_summaries):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

    for gen_summary, ref_summary in zip(generated_summaries, reference_summaries):
        score = scorer.score(ref_summary, gen_summary)
        scores['rouge1'].append(score['rouge1'].fmeasure)
        scores['rouge2'].append(score['rouge2'].fmeasure)
        scores['rougeL'].append(score['rougeL'].fmeasure)

    # Average the scores
    avg_scores = {key: np.mean(val) for key, val in scores.items()}
    return avg_scores
```
